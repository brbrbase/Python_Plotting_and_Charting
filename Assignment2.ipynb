The data for this assignment comes from a subset of The National Centers for Environmental Information (NCEI) [Daily Global Historical Climatology Network](https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/readme.txt) (GHCN-Daily). Each row in the assignment datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)
    * TMIN : Minimum temperature (tenths of degrees C)
* **value** : data value for element (tenths of degrees C)

The data you have been given is near **Louisville, Colorado, United States**, and the stations the data comes from are shown on the map below.

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np
%matplotlib notebook

#Dataset [global variable]
file = 'data/C2A2_data/BinnedCsvs_d400/09b170157ed11c83fccaab1c0e0c12ae13bf082efe27990e5872f8a1.csv'

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'09b170157ed11c83fccaab1c0e0c12ae13bf082efe27990e5872f8a1')
```

```python
#Setup data for plotting
def clean_dataset(file):
    df = pd.read_csv(file)
    df['Data_Value'] = df['Data_Value'].apply(lambda x: x*0.1*9/5 + 32) #convert from tenths deg. C to deg. F
    clean_df = df.sort_values(['ID', 'Date'])  #sort by station ID then by date
    clean_df['Year'], clean_df['Month-Date'] = zip(*clean_df['Date'].apply(lambda x: (x[:4], x[5:])))  #split year and month-date into separate columns
    clean_df = clean_df[clean_df['Month-Date'] != '02-29'] #exclude leap year data
    return clean_df

def calculate_max_min(df):
    min_data = (df[df['Element'] == 'TMIN'].groupby('Month-Date')
                                           .aggregate({'Data_Value': np.min}))
    max_data = (df[df['Element'] == 'TMAX'].groupby('Month-Date')
                                           .aggregate({'Data_Value': np.max}))
    return min_data, max_data
    
#Import, clean and format data
df = clean_dataset(file)
data_10yr = df[df['Year'] != '2015']
data_2015 = df[df['Year'] == '2015']
min_data_10yr, max_data_10yr = calculate_max_min(data_10yr)
min_data_2015, max_data_2015 = calculate_max_min(data_2015)
new_min_record = np.where(min_data_2015['Data_Value'] < min_data_10yr['Data_Value'])[0]
new_max_record = np.where(max_data_2015['Data_Value'] > max_data_10yr['Data_Value'])[0]
```

```python
#Line plot for 2005-2014 data
plt.figure(1)
plt.plot(min_data_10yr.values, 'b', label='Record Low')
plt.plot(max_data_10yr.values, 'r', label='Record High')
plt.gca().axis([0, 365, -50, 115])
plt.xticks(range(0, len(min_data_10yr), 30), min_data_10yr.index[range(0, len(min_data_10yr), 30)], rotation='45')
plt.gca().fill_between(range(len(min_data_10yr)), min_data_10yr['Data_Value'], max_data_10yr['Data_Value'], facecolor = 'orange', alpha = 0.5)
plt.title('Record Temperature Summary\nBoulder, Colorado (2005-2014)')
plt.xlabel('Days of the Year')
plt.ylabel('Temperature (Degrees Fahrenheit)')
plt.legend(loc=8, frameon=False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
#plt.show()
```

```python
#Overlay 2015 data on record min/max data
#Original data plot
plt.figure(2)
plt.plot(min_data_10yr.values, 'b', label='Record Low (2005-2014)')
plt.plot(max_data_10yr.values, 'r', label='Record High (2005-2014)')

#Overlay of 2015 data
plt.scatter(new_min_record, min_data_2015.iloc[new_min_record], s = 20, c = 'black', label = 'New Low Record (2015)', zorder=10)
plt.scatter(new_max_record, max_data_2015.iloc[new_max_record], s = 20, c = 'purple', label = 'New High Record (2015)', zorder=10)

plt.gca().axis([0, 365, -75, 115])
plt.xticks(range(0, len(min_data_10yr), 30), min_data_10yr.index[range(0, len(min_data_10yr), 30)], rotation='45')
plt.gca().fill_between(range(len(min_data_10yr)), min_data_10yr['Data_Value'], max_data_10yr['Data_Value'], facecolor = 'orange', alpha = 0.5)
plt.title('Record Temperature Summary\nBoulder, Colorado')
plt.xlabel('Days of the Year')
plt.ylabel('Temperature (Degrees Fahrenheit)')
plt.legend(loc=8, frameon=False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
